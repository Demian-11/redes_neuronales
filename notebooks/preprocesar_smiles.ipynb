```python
# Ajusta BASE_DIR a tu estructura
# Estando en la carpeta "reto", la carpeta del dataset es "smiles"
BASE_DIR = "/Users/damian/Universidad/redes_neuronales/reto/smiles"

# Las imágenes .ppm están aquí:
IMG_DIR = os.path.join(BASE_DIR, "lfwcrop_color", "faces")

# Los txt con las listas están directamente en smiles/
SMILE_TXT = os.path.join(BASE_DIR, "SMILE_list.txt")
NOT_SMILE_TXT = os.path.join(BASE_DIR, "NON-SMILE_list.txt")

# Carpetas destino para las imágenes separadas
OUT_SMILE = os.path.join(BASE_DIR, "smile")
OUT_NOT_SMILE = os.path.join(BASE_DIR, "not_smile")

os.makedirs(OUT_SMILE, exist_ok=True)
os.makedirs(OUT_NOT_SMILE, exist_ok=True)

def mover_imagenes(lista_txt, destino):
    with open(lista_txt, "r") as f:
        for line in f:
            nombre = line.strip()
            if not nombre:
                continue

            # quitar extensión .jpg del nombre de la lista
            base = os.path.splitext(nombre)[0]
            candidato_ppm = os.path.join(IMG_DIR, base + ".ppm")

            if os.path.exists(candidato_ppm):
                shutil.copy2(
                    candidato_ppm,
                    os.path.join(destino, os.path.basename(candidato_ppm))
                )
            else:
                print(f"No encontré el archivo para: {nombre} -> {candidato_ppm}")

# 1) Copiar .ppm a smile / not_smile
mover_imagenes(SMILE_TXT, OUT_SMILE)
mover_imagenes(NOT_SMILE_TXT, OUT_NOT_SMILE)

def convertir_ppm_a_jpg_y_borrar(carpeta):
    for ppm_path in glob.glob(os.path.join(carpeta, "*.ppm")):
        img = Image.open(ppm_path)
        jpg_path = os.path.splitext(ppm_path)[0] + ".jpg"
        img.save(jpg_path)

        if os.path.exists(jpg_path):
            os.remove(ppm_path)
        else:
            print(f"ERROR: no se generó el JPG para {ppm_path}")

# 2) Convertir a .jpg y borrar .ppm
convertir_ppm_a_jpg_y_borrar(OUT_SMILE)
convertir_ppm_a_jpg_y_borrar(OUT_NOT_SMILE)

print("Listo: carpetas 'smile' y 'not_smile' con imágenes .jpg.")
```
